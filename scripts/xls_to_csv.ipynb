# Skrypt wczytujacy dane .xls i zapisujacy je w formie csv
(Wstępne czysczenie danych)

```python
import pandas as pd


df = pd.read_excel("dane-zeliwo-ADI.xls",skiprows=12)
```

```python
df = df.iloc[:,10:57]
```

```python
kolumny = [df.columns.str.findall('Unnamed')[i] != ['Unnamed'] for i in range(df.shape[1])]
```

```python
df = df.loc[:,kolumny]
```

```python
df.to_csv('dane-zeliwo.csv')
```
